```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/wx_big/training_log.csv"
# training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/wx_pad/training_log.csv"
# training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/unetpp_era5_forcing/training_log.csv"
# training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/deeplabp_pad/training_log.csv"
# training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/deeplabp_era5_forcing/training_log.csv"
# training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/goes_era5_forcing_subset/training_log.csv"
# training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/goes_only/simple_wxformer_lognormal/training_log.csv"
# training_log_path = "/glade/derecho/scratch/dkimpara/goes_10km_train/simple_wxformer/training_log.csv"
df = pd.read_csv(training_log_path)

def plot_metric(metric="train_rmse"):
    columns = sorted([col for col in df.columns if metric in col])
    
    epochs = df["epoch"]
    N = len(columns)

    plt.rcParams["axes.prop_cycle"] = plt.cycler("color", plt.cm.cividis(np.linspace(0,1,N)))

    for k,column in enumerate(columns):
        plt.plot(epochs, df[column], label=column, marker="."
                 #alpha=1/(k+1), c="0"
        )
        plt.axhline(0.3)
    plt.title(metric)
    plt.legend()
    plt.show()

plot_metric()
plot_metric("valid_rmse")
```
