```python
import xarray as xr
import numpy as np
from credit.physics_constants import RAD_EARTH, RVGAS, RDGAS, GRAVITY, RHO_WATER, LH_WATER
import os
```

```python
def get_area(lat, lon):
    lat_rad = np.deg2rad(lat)
    lon_rad = np.deg2rad(lon)
    sin_lat_rad = np.sin(lat_rad)
    d_phi = np.gradient(sin_lat_rad, edge_order=2)
    d_lambda = np.gradient(lon_rad, edge_order=2)[0]
    d_lambda = (d_lambda + np.pi) % (2 * np.pi) - np.pi
    return np.abs(RAD_EARTH**2 * d_phi * d_lambda)
```

```python
for file in ["/glade/campaign/cisl/aiml/wchapman/MLWPS/STAGING/f.e21.CREDIT_climate.statics_1.0deg_32levs_latlon_v3.nc"]:
    filename = os.path.basename(file)
    ds = xr.open_dataset(file)
    area = get_area(ds.latitude, ds.longitude)
    da = xr.DataArray(area, coords={"latitude": ds.latitude}, dims={"latitude": ds.latitude})
    ds["surface_area"] = da
    ds.to_netcdf(os.path.join("/glade/derecho/scratch/dkimpara", filename))
```
