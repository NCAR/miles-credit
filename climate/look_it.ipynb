```python
from scipy.integrate import solve_ivp
import numpy as np

# Lorenz equations (perturbed version)
def lorenz63_perturbed(t, state, sigma, rho, beta, epsilon=0):
    x, y, z = state
    dx = sigma * (y - x)
    dy = x * ((rho + epsilon) - z) - y  # Perturbed rho
    dz = x * y - beta * z
    return [dx, dy, dz]

# Parameters
sigma = 10
rho = 28
beta = 8/3
epsilon = 0.01  # Small perturbation

# Initial condition and time span
initial_state = [1.0, 1.0, 1.0]
t_span = (0, 50)
t_eval = np.linspace(0, 50, 10000)

# Solve the unperturbed system
sol_unperturbed = solve_ivp(lorenz63_perturbed, t_span, initial_state, t_eval=t_eval, args=(sigma, rho, beta, 0))
trajectory_unperturbed = sol_unperturbed.y.T  # Unperturbed trajectory

# Solve the perturbed system
sol_perturbed = solve_ivp(lorenz63_perturbed, t_span, initial_state, t_eval=t_eval, args=(sigma, rho, beta, epsilon))
trajectory_perturbed = sol_perturbed.y.T  # Perturbed trajectory

# Extract z from the trajectories
z_unperturbed = trajectory_unperturbed[:, 2]  # z from the unperturbed system
z_perturbed = trajectory_perturbed[:, 2]      # z from the perturbed system

# Compute time-averaged change in z
delta_z = np.mean(z_perturbed - z_unperturbed)

# Compute susceptibility
susceptibility = delta_z / epsilon
print("Susceptibility:", susceptibility)
```
